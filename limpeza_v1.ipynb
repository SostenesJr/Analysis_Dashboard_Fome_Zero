```python
import pandas as pd
import inflection as ifn
```

# Colete os dados:

```python
dataframe = pd.read_csv("zomato.csv")
```

```python
def rename_columns(dataframe):
    df = dataframe.copy()
    title = lambda x: ifn.titleize(x)
    snakecase = lambda x: ifn.underscore(x)
    spaces = lambda x: x.replace(" ", "")
    cols_old = list(df.columns)
    cols_old = list(map(title, cols_old))
    cols_old = list(map(spaces, cols_old))
    cols_new = list(map(snakecase, cols_old))
    df.columns = cols_new
    return df

df = rename_columns(dataframe)
```

```python
df1 = df.copy()
```

# Limpeza

## Existem colunas que podem ser removidas do DataFrame, por possuir somente 1 valor em todas as linhas?

```python
df1.head()
```

```python
if df1['switch_to_order_menu'].nunique() == 1:
    # Remover a coluna
    df1 = df1.drop(columns=['switch_to_order_menu'])
    print("A coluna foi removida.")
else:
    print("A coluna não contém apenas um valor.")
```

```python
if df1['currency'].nunique() == 1:
    # Remover a coluna
    df1 = df1.drop(columns=['Currency'])
    print("A coluna foi removida.")
else:
    print("A coluna não contém apenas um valor.")
```

```python
# adicionando a coluna Countries
COUNTRIES = {
    1: "India",
    14: "Australia",
    30: "Brazil",
    37: "Canada",
    94: "Indonesia",
    148: "New Zeland",
    162: "Philippines",
    166: "Qatar",
    184: "Singapure",
    189: "South Africa",
    191: "Sri Lanka",
    208: "Turkey",
    214: "United Arab Emirates",
    215: "England",
    216: "United States of America",
}
def country_name(country_id):
    return COUNTRIES[country_id]

df1['countries'] = df1['country_code'].apply(country_name)

```

```python
# tipo de comida
def create_price_tye(price_range):
    if price_range == 1:
        return "cheap"
    elif price_range == 2:
        return "normal"
    elif price_range == 3:
        return "expensive"
    else:
        return "gourmet"
    
df1['type_food'] = df1['price_range'].apply(create_price_tye)
```

```python
COLORS = {
    "3F7E00": "darkgreen",
    "5BA829": "green",
    "9ACD32": "lightgreen",
    "CDD614": "orange",
    "FFBA00": "red",
    "CBCBC8": "darkred",
    "FF7800": "darkred",
}
def color_name(color_code):
    return COLORS[color_code]

df1['color'] = df1['rating_color'].apply(color_name)
```

```python
df1['cuisines'] = df1['cuisines'].astype(str)
```

```python
df1["cuisines"] = df1.loc[:, "cuisines"].apply(lambda x: x.split(",")[0])
```
